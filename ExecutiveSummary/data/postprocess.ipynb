```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
t1 = pd.read_csv("C:\\Users\\alber\\Desktop\\Active_Constraints\\Assessment\\Results\\Training1.csv")
t1dist_A = [t1.loc[np.logical_and(t1["repetition"]==r, t1["assisted"])]["avg_dist"] for r in range(9)]
t1dist_mean_A = [np.mean(r) for r in t1dist_A]
t1dist_std_A = [np.std(r) for r in t1dist_A]
t1dist_N = [t1.loc[np.logical_and(t1["repetition"]==r, t1["assisted"]==False)]["avg_dist"] for r in range(9)]
t1dist_mean_N = [np.mean(r) for r in t1dist_N]
t1dist_std_N = [np.std(r) for r in t1dist_N]
plt.plot([1,2,3,4,5,6,7,8,9],t1dist_mean_N,label='N')
# plt.errorbar([1,2,3,4,5,6,7,8,9],t1dist_mean_N,t1dist_std_N)
plt.plot([1,2,3,4,5,6,7,8,9],t1dist_mean_A,label='A')
# plt.errorbar([1,2,3,4,5,6,7,8,9],t1dist_mean_A,t1dist_std_A)
plt.legend()
plt.show()
```
