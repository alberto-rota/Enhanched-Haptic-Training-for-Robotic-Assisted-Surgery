# Executive Summary Graphs

```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
mpl.rc('font', **{'family': 'serif', 'serif': ['Computer Modern'],'size': 16})
mpl.rc('text', usetex=True)
import numpy as np
```

```python
t = 2
h = 3
w = 1/h
x = np.linspace(0,10,100)
y = 1/(1+np.exp(-1*5*w*(x-t-h)))
```

```python
C2I = 2/2.54
plt.figure(figsize=(8.11*C2I,3*C2I))
plt.plot(x,y)
plt.vlines(t,0,1,'g',linestyles='dashed')
plt.vlines(t+h,0,1,'r',linestyles='dashed')
plt.grid(True)
plt.title("\\textbf{Error Mapping function}")
plt.xlabel("Error [-]")
plt.ylabel("Normalized error [adim.]")
plt.savefig("images/mappingfunction.eps",bbox_inches='tight',dpi=300)
plt.show()
```
