```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
mpl.rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
mpl.rc('text', usetex=True)
import numpy as np
```

```python

dfun = pd.read_csv("C:\\Users\\alber\\Desktop\\Suturing_0\\Suturing_0_VFs.csv",index_col=None)
dfas = pd.read_csv("C:\\Users\\alber\\Desktop\\Suturing_1\\Suturing_1_VFs.csv",index_col=None)

plt.figure(figsize=(10,3))
plt.plot(dfas['Time']-dfas['Time'][0],dfas['TrajectoryOrientationGuidanceVFRL_dist0']*1000,label='Assisted')
plt.plot(dfun['Time']-dfun['Time'][0],dfun['TrajectoryOrientationGuidanceVFRL_dist0']*1000,label='Unassisted')
plt.legend()
plt.grid()

plt.xlabel("Time [s] since beginning of the task")
plt.ylabel("Error [mm]")
plt.title("\\bf{Distance from preop-planned trajectory}")
plt.show()
```
